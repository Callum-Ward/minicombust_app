```python
import matplotlib.pyplot as plt
import numpy as np
import re
import pandas as pd
import csv
```

```python
weak_particles_df = pd.read_csv("out/tx2-weak_particles-1nodes-64ppn-80cells_modifier-emitted_particles.log")
print(weak_particles_df)
```

```python
runtimes_file=open("../out/runtimes.log", 'r')

num_particles = []
times         = []

for line in runtimes_file:
    split_line = re.split('/|_|\n| ', line)
    num_particles.append(int(split_line[1]))
    times.append(float(split_line[4][:-1]))

x = np.array(num_particles)
y = np.array(times)
p = x.argsort()
x=x[p]
y=y[p]

plt.subplots(figsize=(10, 5))
plt.title("Runtime vs Num Particles (1000 timesteps)")
plt.xlabel("Number of particle emitted per timestep")
plt.ylabel("Runtime (s)")
plt.grid()
plt.xlim(0, max(num_particles)+2)
# plt.ylim(0, max(num_particles)+0.2);
plt.ylim(0, max(times)+0.2);


plt.plot(x, y)
```

```python
runtimes_file=open("../out/avg_num_particle.log", 'r')

avg_particles = []

for line in runtimes_file:
    split_line = re.split(':', line)
    avg_particles.append(float(split_line[2][:-1]))

avg_particles = np.array(avg_particles)
avg_particles=avg_particles[p]

plt.subplots(figsize=(10, 5))
plt.title("Runtime vs Avg Particles (1000 timesteps)")
plt.xlabel("Avg Particles in Timestep")
plt.ylabel("Runtime (s)")
plt.grid()
plt.xlim(0, max(avg_particles)+2)
# plt.ylim(0, max(num_particles)+0.2);
plt.ylim(0, max(times)+0.2);


plt.plot(avg_particles, y)
```

```python
files = ["emitted.log",
         "interpolate_nodal_data.log",
         "particle_interpolation_data.log",
         "solve_spray_equations.log",
         "update_particle_positions.log"]

for f in files:
    print("../out/" + f)
    curr_file=open("../out/" + f, 'r')
    
    num_particles = []
    times         = []    
    bandwidths    = []



    for line in curr_file:
        split_line = re.split('/|\n| |:', line)
        fname = split_line[1]
        
        num_particles.append(int(fname.split('_')[0]))
        times.append(float(split_line[5]))
        bandwidths.append(float(split_line[8]))


    x = np.array(num_particles)
    y = np.array(times)
    z = np.array(bandwidths)
    p = x.argsort()
    x=x[p]
    y=y[p]
    z=z[p]

    plt.subplots(figsize=(10, 5))
    plt.title(f[:-4] + " Runtime vs Num Particles (1000 timesteps) ")
    plt.xlabel("Number of particle emitted per timestep")
    plt.ylabel("Runtime (s)")
    plt.grid()
    plt.xlim(0, max(num_particles)+0.1*max(num_particles))
    plt.ylim(0, max(times)+0.1*max(times));
    plt.plot(x, y)
    
    plt.subplots(figsize=(10, 5))
    plt.title(f[:-4] + " Bandwidth vs Num Particles (1000 timesteps)")
    plt.xlabel("Number of particle emitted per timestep")
    plt.ylabel("Bandwidth (GB/s)")
    plt.grid()
    plt.xlim(0, max(num_particles)+0.1*max(num_particles))
    plt.ylim(0, max(bandwidths)+0.1*max(bandwidths));
    plt.plot(x, z)
    
```
